```python
import pandas as pd
import matplotlib.pyplot as plt

# Use pd.read_csv() to read the CSV file into a DataFrame
df_routes = pd.read_csv='/Users/anna-lenaschwarz/Desktop/Adpro/group_03/downloads/routes.csv'

def distance_analysis(flghts_df, bins=30, show_mean=True, show_median=True):
    """
    Plots the distribution of flight distances.

    Parameters:
    - flights_df: DataFrame containing flight distances.
    - bins: Number of bins for the histogram.
    - show_mean: Whether to show the mean distance line.
    - show_median: Whether to show the median distance line.
    """
    plt.figure(figsize=(10, 6))
    
    # Plotting the histogram
    plt.hist(flights_df['distance'], bins=bins, color='skyblue', edgecolor='black', alpha=0.7)
    
    # Optionally adding mean and median lines
    if show_mean:
        mean_val = flights_df['distance'].mean()
        plt.axvline(mean_val, color='red', linestyle='dashed', linewidth=1, label=f'Mean: {mean_val:.2f}')
    
    if show_median:
        median_val = flights_df['distance'].median()
        plt.axvline(median_val, color='green', linestyle='dashed', linewidth=1, label=f'Median: {median_val:.2f}')
    
    # Adding labels and title
    plt.xlabel('Flight Distance')
    plt.ylabel('Frequency')
    plt.title('Distribution of Flight Distances')
    plt.legend()
    
    # Show plot
    plt.show()

distance_analysis(flights_df, bins=10)  # Adjust bins as needed
```
